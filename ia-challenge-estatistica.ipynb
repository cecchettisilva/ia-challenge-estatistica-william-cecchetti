## Exercício do módulo
Com base nas aulas do módulo de Estatística, utilize os métodos da biblioteca Pandas para tratar o dicionário abaixo e:
- Trazer a média das vendas
- Criar um gráfico de barras vertical mostrando o mês de referência e o valor
- Criar um gráfico de linhas mostrando o mês de referência e o valor

```python
# Importando bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Validando de que as bibliotecas estão instaladas
pd.__version__
```

```python
# Dicionário de faturamento para criação do DataFrame
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
    ]
}
```

```python
# Validando que o dicionário foi criado corretamente
dict_faturamento
```

```python
# Criando o DataFrame a partir do dicionário
dict_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
# Validando criação do DataFrame a partir do dicionário
dict_faturamento
```

### Média de vendas

```python
# Trazer a média das vendas
dict_faturamento["valor"].mean()
```

### Criar um gráfico de barras vertical mostrando o mês de referência e o valor

```python
# Gráfico de barras com mês de referência e valor de faturamento
ax = dict_faturamento.plot.bar(x='data_ref', y='valor', color='purple', figsize=(12, 6))

# Adicionando título e rótulos
ax.set_title('Faturamento por Mês', fontsize=16)
ax.set_xlabel('Mês de Referência', fontsize=12)
ax.set_ylabel('Faturamento (R$)', fontsize=12)
```

### Criar um gráfico de linhas mostrando o mês de referência e o valor

```python
# Gráfico de linhas com mês de referência e valor de faturamento
ax = dict_faturamento.plot.line(x='data_ref', y='valor', color='purple', figsize=(12, 6))

# Adicionando título e rótulos
ax.set_title('Faturamento por Mês', fontsize=16)
ax.set_xlabel('Mês de Referência', fontsize=12)
ax.set_ylabel('Faturamento (R$)', fontsize=12)
```
